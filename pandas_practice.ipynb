```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
df = pd.read_csv('pokemon_data.csv')
df.columns
df.index = df['#']
df.drop(['#'],axis=1)
```

```python
df[['Type 2']].value_counts()
```

```python
df.describe()
```

```python
df= df.rename(columns = {'Type 1':'type1', 'Type 2': 'type2'})
```

```python
## check the mean statistics for all types of pokemons
df.groupby(["type1"])[["Speed","Attack","HP","Defense"]].mean()
```

```python
df.groupby(["type1"])[["Speed","Attack","HP","Defense"]].agg([np.mean,max,min])
```

```python
df.pivot_table(values='Speed',index='type1').sort_values('Speed')
### FLYING HAVE THE MOST SPEED on an Average and Fairy has the least Speed on average
```

```python

### IN LEGENDARY
##  FLYING POKEMONS HAVE THE MOST SPEED
### ATTACK is best for Ground Legendary pokemons
### ROCK HAVE THE BEST DEFENCE
df[df.loc[:,['type1','Speed','Legendary']]['Legendary']==True].groupby(['type1'])[['Speed','Attack','Defense']].mean().sort_values(by='Defense')

```

```python
legendary = df[df['Legendary']==True]
df[df['Legendary']==True]['HP'].hist(alpha=0.7,label='Legendary')
df[df['Legendary']==False]['HP'].hist(alpha=0.7,label='NL')
```
